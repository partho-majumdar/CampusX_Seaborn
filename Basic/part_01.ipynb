```python
'''' 
------------------------ Why Seaborn
    - Easy to use 
    - better aesthetics
    - more graph include

Types of function in seaborn
    - Figure level
    - Axis level

    ----------------------------
    -                          -
    -                          -
    -                          -
    -                          -
    -                          -
    -                          -
    ----------------------------

    this container = figure
    inside this container if i draw a graph is known as = axis

    top (figure)
        - multiple axis
            - levels, line, marker, xticks, yticks

            
----------> Main classification <----------
    - Relational Plot (Scatter, Line)
    - Distribution Plot (Histogram, KDE, Rug)
    - Categorical Plot (Bar, Count, Box, Violin, Swam)
    - Matrix Plot (Heatmap, Cluster-map)
    - Regression Plot ()
    - Multi-plots (Joint, Pair)
'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import scipy
```

```python
# ------------------ Relational Plot
''' 
- To see the statistical relation between 2 or more variable
- Bivariate analysis
'''

tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot (axis level function) -> numerical data
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# relplot (figure level function) -> square shape (recommended) cause using this i can plot multiple figure level plot
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter')
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'line')
```

```python
# in scatter plot i want each dot male or female (hue parameter --> different color)
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'sex')
```

```python
# how many people came in lunch and dinner (using style)
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'time')
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'sex', style = 'time')
```

```python
# each customer come with how many customer 
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'time', size = 'size')
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'sex', style = 'time', size = 'size')
```

```python
# line plot (if i want to show continuity)
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
```

```python
sns.lineplot(data = temp_df, x = 'year', y = 'lifeExp', hue = 'country')
sns.relplot(data = temp_df, x = 'year', y = 'lifeExp', kind = 'line', hue = 'country')
```

```python
# i want to see lifeExp of 3 country
temp_df = gap[gap['country'].isin(['India', 'Brazil', 'Germany'])]

 # figure level function so legend are outside graph
sns.relplot(kind = 'line', data = temp_df, x = 'year', y = 'lifeExp', hue = 'country')
```

```python
# this is axis level function hence legend are inside graph
sns.lineplot(data = temp_df, x = 'year', y = 'lifeExp', hue = 'country') 
```

```python
# i want to show which country are from which continental
sns.relplot(kind = 'line', data = temp_df, x = 'year', y = 'lifeExp', hue = 'country', style = 'continent')
```

```python
# size (different level of thickness)
sns.relplot(kind = 'line', data = temp_df, x = 'year', y = 'lifeExp', hue = 'country', style = 'continent', size = 'continent')
```

```python
# facet plot (using particular categorical column i can make multiple plot)
''' 
    - This is figure level function
    - work with relplot
    - will not work with scatterplot, lineplot
'''
sns.relplot(kind = 'scatter', data = tips, x = 'total_bill', y = 'tip', hue = 'sex')

# plot 2 graph one for male and another for female
sns.relplot(kind = 'scatter', data = tips, x = 'total_bill', y = 'tip', col = 'smoker')
sns.relplot(kind = 'scatter', data = tips, x = 'total_bill', y = 'tip', row = 'sex')
```

```python
sns.relplot(kind = 'scatter', data = tips, x = 'total_bill', y = 'tip', row = 'sex', col = 'smoker')
```

```python
sns.relplot(kind = 'scatter', data = tips, x = 'total_bill', y = 'tip', row = 'day', col = 'sex')
```

```python
sns.relplot(kind = 'line', data = tips, x = 'total_bill', y = 'tip', row = 'day', col = 'sex')
```

```python
# i want lifeExp vs gdpPercap scatterplot
sns.relplot(data = gap, kind = 'scatter', x = 'lifeExp', y = 'gdpPercap')
```

```python
# for each year i want different graph
# col_wrap -> split column and send it to next row
sns.relplot(kind = 'scatter', data = gap, x = 'lifeExp', y = 'gdpPercap', col = 'year', col_wrap = 3)
```

```python
# ------------------------- Distribution Plot
''' 
This plot tell me how my data is divided 
    - Used for univariate analysis
    - Range of observation
    - Central tendency 
    - is data bimodal (multiple pick)
    - are there outliers

    figure level --> dis plot
    axis level --> hist plot / kde plot / rug plot
'''
```

```python
# plotting univariate histogram
sns.histplot(data = tips, x = 'total_bill') 
```

```python
# using dis plot
sns.displot(data = tips, kind = 'hist', x = 'total_bill')
```

```python
# editing bins
sns.displot(data = tips, kind = 'hist', x = 'total_bill', bins = 20) # total bins divide into 20 equal bins
```

```python
# generally histogram use in numerical data but i can use it in categorical data
sns.displot(data = tips, kind = 'hist', x = 'day')
```

```python
# how 2 different histogram look like (hue parameter)
sns.displot(data = tips, kind = 'hist', x = 'tip', hue = 'sex')
```

```python
# for clear visualization -> element = 'step'
sns.displot(kind = 'hist', data = tips, hue = 'sex', x = 'tip', element = 'step')
```

```python
titanic = sns.load_dataset('titanic')

# plot histogram of age 
sns.displot(kind = 'hist', data = titanic, x = 'age', element = 'step', hue = 'sex')
```

```python
# facet col and row --> not work in histplot function cause it is axis level function 
sns.displot(data = tips, kind = 'hist', x = 'total_bill', element = 'step', col = 'sex')
```

```python
# kde plot --> kernel density estimation  
'''
    - it not use bins and not return count
    - use gaussian kernel function and make data smooth (this is kde)
'''

# axis level function 
sns.kdeplot(data = tips, x = 'total_bill')
```

```python
# figure level function
sns.displot(data = tips, kind = 'kde', x = 'total_bill')
```

```python
# different color for male and female
sns.displot(data = tips, kind = 'kde', x = 'total_bill', hue = 'sex')
```

```python
# fill kde plot
sns.displot(data = tips, kind = 'kde', x = 'total_bill', hue = 'sex', fill = True)
```

```python
# rug plot 
''' 
    - plot marginal distribution by drawing ticks along the x and y axis  
'''

sns.kdeplot(data = tips, x = 'total_bill')
sns.rugplot(data = tips, x = 'total_bill')
```

```python
# Bivariate histogram 

sns.histplot(data = tips, x = 'total_bill', y = 'tip') 
sns.displot(data = tips, x = 'total_bill', y = 'tip', kind = 'hist')
```

```python
# Bivariate kde plot --> smooths the (x, y) observation with a 2D Gaussian
sns.kdeplot(data = tips, x = 'total_bill', y = 'tip')
sns.displot(data = tips, kind = 'kde', x = 'total_bill', y = 'tip')
```

```python
# ---------------------- Matrix Plot
''' 
    - Heat map --> Plot rectangular data in colored encode matrix 
    - Cluster map
'''
```

```python
# heat map --> axis level function
temp_df = gap.pivot(index = 'country', columns = 'year', values = 'lifeExp')
plt.figure(figsize = (15, 15))
sns.heatmap(temp_df)
```

```python
# same graph for europe country (not color also lifeExp value) --> annot =True
temp_df = gap[gap['continent'] == 'Europe'].pivot(index = 'country', columns = 'year', values = 'lifeExp')
plt.figure(figsize = (15, 15))
sns.heatmap(temp_df, annot = True)
```

```python
# make a gap in graph using linewidth
plt.figure(figsize = (15, 15))
sns.heatmap(temp_df, annot = True, linewidths = 0.5, cmap = 'viridis')
```

```python
# clustermap --> internal use cluster (make different box for similar and dissimilar value)
iris = px.data.iris()
```

```python
# find similarity from iris dataset and which flower are similar to each other
sns.clustermap(iris.iloc[:, [0, 1, 2, 3]]) # internally it perform (Agglomerative Hierarchical Clustering) ML model
```
