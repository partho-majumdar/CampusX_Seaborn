```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import scipy
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
# ----------------------- Categorical Plot
''' 
    - Categorical scatter plot (strip-plot, swarm-plot)
    - Categorical distribution plot (box-plot, violin-plot)
    - Categorical estimate plot (bar-plot, point-plot, count-plot)
'''
```

```python
# -------------------- Categorical scatter plot (Bivariate analysis)

# strip plot (pick one categorical col, pick one numerical col and then plot scatter plot)
# axis level function
sns.stripplot(data = tips, x = 'day', y = 'total_bill')
```

```python
sns.stripplot(data = tips, x = 'day', y = 'total_bill', jitter = False) # everything lie on one column
```

```python
# figure level function (catplot)
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'strip')
```

```python
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'strip', jitter = 0.2, hue = 'sex') # point are 
```

```python
# swarmplot (kind of strip plot but here internally run algorithm that help how my data is divided) -> better representation
# better for small dataset
# figure level function
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'swarm')
```

```python
# axis level function 
sns.swarmplot(data = tips, x = 'day', y = 'total_bill')
```

```python
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'swarm', hue = 'sex')
```

```python
# --------------------------- Categorical distribution plot (Univariate analysis)

# boxplot
''' 
- give 5 number summery (Min, First quartile, Median, Third quartile, Max)
Minimum outlier (Q1 - 1.5IQR) --------------- IQR (Q1 - Q3) -------------- Maximum outlier (Q1 + 1.5IQR)
'''
```

```python
# axis level function
sns.boxplot(data = tips, x = 'sex', y = 'total_bill')
```

```python
sns.boxplot(data = tips, x = 'day', y = 'total_bill')
```

```python
# figure level function
sns.catplot(data = tips, kind = 'box', x = 'day', y = 'total_bill')
```

```python
sns.catplot(data = tips, kind = 'box', x = 'day', y = 'total_bill', hue = 'sex')
```

```python
# single boxplot (numeric column)
sns.boxplot(data = tips, y = 'total_bill')
```

```python
# violinplot (boxplot + kdeplot)
# axis level function
sns.violinplot(data = tips, x = 'day', y = 'total_bill')
```

```python
# figure level function
sns.catplot(data = tips, x = 'day', y = 'total_bill', kind = 'violin')
```

```python
# hue parameter
sns.catplot(kind = 'violin', data = tips, x = 'day', y = 'total_bill', hue = 'sex')
```

```python
# if i want 2 side by side plot (using split = True)
sns.catplot(kind = 'violin', data = tips, x = 'day', y = 'total_bill', hue = 'sex', split = True)
```

```python
# ------------------ Categorical Estimate plot (measure central tendency) 
```

```python
# barplot (numerical avg vs categorical)
# axis level function
sns.barplot(data = tips, x = 'sex', y = 'total_bill')

# here | bar is known as error bar (when x axis value is repeat) -> use bootstrapping to compute a confidence interval around the estimate
```

```python
# remove error bar
# sns.catplot(kind = 'bar', data = tips, x = 'sex', y = 'total_bill', errorbar = None)
sns.barplot(data = tips, x = 'sex', y = 'total_bill', errorbar = None)
```

```python
# hue parameter
sns.catplot(data = tips, kind = 'bar', x = 'sex', y = 'total_bill', hue = 'smoker', estimator = np.max)
```

```python
# pointplot (basically a barplot just connect the point of errorbar) 
sns.pointplot(data = tips, x = 'sex', y = 'total_bill')
```

```python
sns.pointplot(data = tips, x = 'sex', y = 'total_bill', hue = 'smoker')
```

```python
sns.pointplot(data = tips, x = 'sex', y = 'total_bill', hue = 'smoker', estimator = np.max)
```

```python
# countplot (show number of observation in each category)
# how many male and female
sns.countplot(data = tips, x = 'sex')
```

```python
# how many male of female come in each day
sns.catplot(kind = 'count', data = tips, x = 'sex', hue = 'day')
```

```python
# faceting using catplot   
sns.catplot(kind = 'box', data = tips, x = 'sex', y = 'total_bill', col = 'smoker', row = 'time')
```

```python
# ----------------- Reg Plot (Regression plot)
# plot scatter plot also plot linear regression line    
```

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
sns.regplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# lmplot (same as regplot also add hue)
sns.lmplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
```

```python
# residplot (when i do regression analysis for some point i got some error) --> error for my prediction
sns.residplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# ------------------ Multi-plot 
```

```python
# figure level -> relplot -> displot -> catplot (everyone has facet)
sns.catplot(data = tips, x = 'sex', y = 'total_bill', kind = 'violin', col = 'day', row = 'time')
```

```python
# FacetGrid 
g = sns.FacetGrid(data = tips, col = 'day', row = 'time', hue = 'smoker')
g.map(sns.boxplot, 'sex', 'total_bill')
g.add_legend()
```

```python
# PairGrid vs PairPlot (plot graph of numerical data of every combination)
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue = 'species')
```

```python
# PairGrid
g = sns.PairGrid(data = iris)
# g.map(sns.histplot)
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(iris, hue = 'species')
g.map(sns.scatterplot)
g.add_legend()
```

```python
# edit diagonal and off-diagonal
g = sns.PairGrid(data = iris, hue = 'species')
# g.map_diag(sns.violinplot)
# g.map_diag(sns.histplot)
g.map_diag(sns.boxplot)
# g.map_offdiag(sns.scatterplot)
g.map_offdiag(sns.kdeplot)
```

```python
# different graph of diagonal upper and diagonal lower
g = sns.PairGrid(data = iris, hue = 'species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# if i want to perform same thing with specific columns use "vars"
g = sns.PairGrid(data = iris, hue = 'species', vars = ['sepal_width', 'petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# JointPlot
sns.jointplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
# sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde')
# sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'reg')
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hist', hue = 'sex')
```

```python
# JointGrid
g = sns.JointGrid(data = tips, x = 'total_bill', y = 'tip')
# g.plot(sns.scatterplot, sns.histplot)
# g.plot(sns.scatterplot, sns.boxplot)
# g.plot(sns.scatterplot, sns.violinplot)
g.plot(sns.kdeplot, sns.violinplot)
```
